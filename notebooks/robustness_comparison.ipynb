```python
import matplotlib.pyplot as plt

labels = ['Clean', 'Noise', 'Occlusion']
accs = {
    'baseline': [98, 9, 98], 
    'randomcrop': [97, 8, 97],
    'colorjitter': [97.2, 10, 96.5],
    'augmix': [97.5, 25, 97],
    'resnet': [91.3, 6.5, 90.9]
}

x = range(len(labels))
width = 0.18

plt.figure(figsize=(9, 5))
for i, (k, v) in enumerate(accs.items()):
    plt.bar([xi + i * width for xi in x], v, width=width, label=k)

plt.xticks([xi + (len(accs) / 2 - 0.5) * width for xi in x], labels)
plt.ylim(0, 100)
plt.ylabel("Accuracy (%)")
plt.title("Robustness Comparison Across Models (EfficientNet + ResNet)")
plt.legend()
plt.tight_layout()
plt.savefig("robustness_comparison_all_models.png")
plt.show()
```
