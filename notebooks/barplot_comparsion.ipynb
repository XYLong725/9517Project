```python
import matplotlib.pyplot as plt
import numpy as np

models = ['SIFT+BoW+SVM', 'EfficientNet-B0', 'EfficientNet+AugMix']
accuracies = [0.52, 0.97, 0.978]
f1_scores = [0.50, 0.97, 0.97]
times = [1014, 225, 208]

x = np.arange(len(models))
width = 0.25

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(x - width, accuracies, width, label='Accuracy')
ax1.bar(x, f1_scores, width, label='F1 Score')
ax1.set_ylabel('Score')
ax1.set_ylim(0, 1.1)
ax1.set_title('Model Comparison: Accuracy vs F1')
ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(x + width, times, 'o--', label='Training Time (s)', color='orange')
ax2.set_ylabel('Training Time (s)')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
