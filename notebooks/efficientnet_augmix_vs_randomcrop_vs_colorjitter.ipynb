```python
import numpy as np
import matplotlib.pyplot as plt

names = ['randomcrop', 'colorjitter', 'augmix']
for name in names:
    acc = np.load(f"acc_{name}.npy")
    plt.plot(acc, label=name)
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy Comparison')
plt.legend()
plt.show()
```
