```python
import numpy as np
import matplotlib.pyplot as plt

loss_files = {
    'baseline': 'loss_baseline.npy',
    'randomcrop': 'loss_randomcrop.npy',
    'colorjitter': 'loss_colorjitter.npy',
    'augmix': 'loss_augmix.npy',
    'resnet18': 'loss_resnet18.npy'
}

methods = list(loss_files.keys())

plt.figure(figsize=(8, 5))
for name in methods:
    loss = np.load(loss_files[name])
    plt.plot(loss, label=name)

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('training_loss_comparison.png')
plt.show()

```
