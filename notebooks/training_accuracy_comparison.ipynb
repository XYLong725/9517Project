```python
import numpy as np
import matplotlib.pyplot as plt

methods = ['baseline', 'randomcrop', 'colorjitter', 'augmix', 'resnet18']

acc_files = {
    'baseline': 'acc_baseline.npy',
    'randomcrop': 'acc_randomcrop.npy',
    'colorjitter': 'acc_colorjitter.npy',
    'augmix': 'acc_augmix.npy',
    'resnet18': 'acc_resnet18.npy'
}

plt.figure(figsize=(8, 5))
for name in methods:
    acc = np.load(acc_files[name])
    plt.plot(acc, label=name)

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy Comparison')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('training_accuracy_comparison.png')
plt.show()
```
