```python
import matplotlib.pyplot as plt
import numpy as np

methods = ['randomcrop', 'colorjitter', 'augmix','resnet']
accs = []
f1s = []

for m in methods:
    with open(f"score_{m}.txt") as f:
        lines = f.readlines()
        accs.append(float(lines[0].split(':')[1]))
        f1s.append(float(lines[1].split(':')[1]))

x = np.arange(len(methods))
width = 0.35

plt.bar(x - width/2, accs, width, label='Accuracy')
plt.bar(x + width/2, f1s, width, label='Macro F1')
plt.xticks(x, methods)
plt.ylabel('Score')
plt.title('Model Comparison by Accuracy and F1')
plt.legend()
plt.show()
```
